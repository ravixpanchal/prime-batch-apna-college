### **Web Scraping**

```python
#Web Scraping
import requests

URL="https://www.scrapethissite.com/pages/simple/"

res=requests.get(URL)
```

```python
if res.status_code==200:
    print(res.headers)

with open("scraped_data/data.html","w",encoding="utf-8") as f:
    f.write(res.text)

    
```

```python
#BeautifulSoup
from bs4 import BeautifulSoup
with open("scraped_data/data.html","r",encoding="utf-8")as f:
    html_data=f.read()
```

```python
soup=BeautifulSoup(html_data,"lxml")
```

```python
all_h3=soup.find_all("h3")

all_countries=[]

for h3 in all_h3:
    country_name=h3.get_text(strip=True)
    # print(h3.find_parent("div"))
    population_data=h3.find_next("div").select("span.country-population")[0].get_text(strip=True)
    # print(h3.find_next("div").select_one("span.country-population").get_text())
    # print(h3.attrs)
    all_countries.append([country_name,population_data])
```

```python
all_countries
```

```python
import pandas as pd
df=pd.DataFrame(all_countries,columns=["Country","Population"])
df
```

```python
df.to_csv("cleaned_data/clean_data.csv", index=False)
    
```
