```python
#Real Data

#oscar winner movies
oscar_movies=[
    "The dark Knight",
    "The Hurt Locker",
    "The King's Speech",
    "The Artist",
    "Argo"
]

years=[2008,2009,2010,2011,2012]
oscar_revenue=[1005,170,427,133,232]

plt.plot(years,oscar_revenue)
plt.title("Oscer movies")
plt.xlabel("Years")
plt.ylabel("Revenue")
plt.show()
```

```python
#non oscar movies revenue
non_oscar_movies=["Slumdog Millionaire","Avatar","Inception","Hugo","Lincoln"]
non_oscar_revenue=[378,2788,829,185,275]
plt.plot(years,oscar_revenue,"o--r", label="oscar movies")
plt.plot(years,non_oscar_revenue,"<-b" ,label="non-oscar movies")
plt.title("Oscar Movies Vs Non-oscar movies revenue")
plt.xlabel("Years")
plt.ylabel("Revenue (in $)")
plt.legend()
plt.show()
```

```python
#non oscar movies revenue

plt.plot(years,oscar_revenue, label="oscar movies")
plt.plot(years,non_oscar_revenue,label="non-oscar movies")
plt.title("Oscar Movies Vs Non-oscar movies revenue")
plt.xlabel("Years")
plt.ylabel("Revenue (in $)")
plt.legend()


plt.style.available
# plt.style.use("default")
plt.grid()
plt.show()
```

```python
#Bar charts
#movies-revenue
movies=[
    "The dark Knight",
    "The Hurt Locker",
    "The King's Speech",
    "The Artist",
    "Argo"
]

years=[2008,2009,2010,2011,2012]
oscar_revenue=[1005,170,427,133,232]

plt.bar(years,oscar_revenue,color="purple")
plt.title("Revenue in each year for oscar movies")
plt.xlabel("Years")
plt.ylabel("Revenue (in $M)")
plt.show()

for i in range(len(years)):
    plt.text(years[i],oscar_revenue[i],str(oscar_revenue[i]),ha="center")
```

```python
import numpy as np
oscar_movies=[
    "The dark Knight",
    "The Hurt Locker",
    "The King's Speech",
    "The Artist",
    "Argo"
]

oscar_revenue=[1005,170,427,133,232]

non_oscar_movies=["Slumdog Millionaire","Avatar","Inception","Hugo","Lincoln"]
non_oscar_revenue=[378,2788,829,185,275]

years=[2008,2009,2010,2011,2012]

x=np.arange(len(years))
width=0.4

plt.bar(x-width/2,oscar_revenue,width,label="Oscar Movies")
plt.bar(x+width/2,non_oscar_revenue,width,label="Non-Oscar Movies")
plt.title("Oscar Vs Non-Oscar Movies Revenue")
plt.xlabel("Years")
plt.ylabel("Revenue(in $)")
plt.legend()
plt.xticks(x,years)
```

```python
#Scatter Plot

people=["People A","People B","People C","People D","People E","People F","People G","People H","People I","People J"]
age=[22,25,30,35,40,45,50,55,60,65]
bp=[110,115,120,122,125,130,135,123,145,150]
colors=["green" if x<135 else "red" for x in bp]
plt.scatter(age,bp,s=bp,cmap="OrRd",c=bp)
plt.title("Age VS BP")
plt.xlabel("Age")
plt.ylabel("BP")
plt.grid()
plt.colorbar(label="BP")

for i in range(len(people)):
    plt.annotate(people[i],xy=(age[i],bp[i]),xytext=(age[i]+1,bp[i]+1))

plt.xlim(min(age),max(age)+10)
plt.ylim(min(bp),max(bp)+10)

```

```python
#Multiple Plots
cities=["City A","City B","City C","City D","City E"]

#Winter: Temp(C) vs Humidity(%)
winter_temp=[5,2,10,0,7]
winter_humidity=[80,75,65,85,70]

#Summer: Temp(C) vs Humidity(%)
summer_temp=[25,30,28,35,27]
summer_humidity=[60,50,55,45,65]

plt.scatter(winter_temp,winter_humidity, label="Winter")
plt.scatter(summer_temp,summer_humidity, label="Summer")
plt.title("Summer VS Winter Data")
plt.xlabel("Temp")
plt.ylabel("Humidity")
plt.legend()
plt.grid()
plt.show()
```
