### Copy Vs View

Slicing a list returns a copy but slicing a numpy array returns a view

```python
#sliced list  is a copy
py_list=[1,2,3,4,5]
copy_list=py_list[1:4] #2,3,4
copy_list[1]=100

print(py_list)
print(copy_list)

#Sliced array is a view
arr=np.array([1,2,3,4,5])
view_array=arr[1:4] #2,3,4
view_array[1]=100

print(arr)
print(view_array)

#Creating a copy of array
copy_arr=arr[1:4].copy()
copy_arr[1]=500
print(copy_arr)
print(arr)
```

```python
#normalize- Z score
arr=np.array([[1,2],[3,4]])
mean=np.mean(arr)
std_dev=np.std(arr)

normalized_arr=(arr-mean)/std_dev

print("mean is : ",mean)
print("standard deviation is : ",std_dev)
print("normalized arr is : \n",normalized_arr)
```
