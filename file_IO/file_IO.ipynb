Word Search


```python
data=True
word="store"
line=1
with open("sample.txt","r") as f:
    while data:
        data=f.readline()
        
        
        if(word in data):
            print(f"{word} found in line {line}")
            break
        # print(data)
        line+=1
        
```

Exceptional Handling

```python
try:
    x=int(input("Enter x: "))
    print(10/x)
    
except ZeroDivisionError:
    print("Don't divide by zero!")
    
except ValueError:
    print(f"invalid input")
else:
    print("Execution Successfully")
finally:
    print("Happy New Year 2026!")
    
```
