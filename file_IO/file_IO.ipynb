Word Search


```python
data=True
word="store"
line=1
with open("sample.txt","r") as f:
    while data:
        data=f.readline()
        
        
        if(word in data):
            print(f"{word} found in line {line}")
            break
        # print(data)
        line+=1
        
```

Exceptional Handling

```python
try:
    x=int(input("Enter x: "))
    print(10/x)
    
except ZeroDivisionError:
    print("Don't divide by zero!")
    
except ValueError:
    print(f"invalid input")
else:
    print("Execution Successfully")
finally:
    print("Happy New Year 2026!")
    
```

```python
#without list comprehension
l=[]
for i in range(5):
    l.append(i*i)
print(l)
```

```python
#with list comprehension
sq=[i*i for i in range(5)]
print(sq)
```

```python
nums=[-2,-3,-5,2,7,8,-8,-6]

nums=[0 if val<0 else val for val in nums]
print(nums)
```

```python
words=["hello","python","apna college"]
words=[val.upper() for val in words]
print(words)
```

Json 


```python
#convert json to string
import json

json_str='{"name":"Ravi Panchal","isTeacher":true}'
py_obj=json.loads(json_str)
print(type(py_obj),py_obj)
```

```python
#convert string to json
import json

py_obj={
    "name":"Ravi Panchal",
    "isTeacher":True,
    "isSalary":None
}

json_str=json.dumps(py_obj)

print(type(json_str),json_str)
```

```python
#directly open with json file

import json

with open("data.json","r") as f:
    py_obj=json.load(f)
    print(py_obj)
    print(type(py_obj))
```

```python
#directly write in the json file

import json

data={
    "name":"Ravi Panchal",
    "age":"22",
    "isTeacher":True
}

with open("data.json","w") as f:
    json_data=json.dump(data,f,indent=5,sort_keys=True)
    
    
```
